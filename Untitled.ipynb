```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your data into a DataFrame
# Replace 'your_data.csv' with the path to your data file
data = pd.read_csv('final_data.csv')

# Extract the "Pressure | (atm)" column
pressure_data = data['Pressure | (atm)']

# Create a boxplot for the original data
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.boxplot(x=pressure_data)
plt.title('Original Data')
plt.xlabel('Pressure | (atm)')

# Calculate IQR
Q1 = pressure_data.quantile(0.25)
Q3 = pressure_data.quantile(0.75)
IQR = Q3 - Q1

# Define thresholds for identifying outliers
lower_threshold = Q1 - 1.5 * IQR
upper_threshold = Q3 + 1.5 * IQR

# Identify outliers
outliers = (pressure_data < lower_threshold) | (pressure_data > upper_threshold)

# Remove outliers
cleaned_data = data[~outliers]

# Create a boxplot for the cleaned data
plt.subplot(1, 2, 2)
sns.boxplot(x=cleaned_data['Pressure | (atm)'])
plt.title('Cleaned Data')
plt.xlabel('Pressure | (atm)')

plt.tight_layout()
plt.show()

cleaned_data.info()
```
